# RPN Calculator Code

```python
import operator
```

```python
def RPN(Input):
    #Define operators
    ops = {'+': operator.add, '-': operator.sub, '*': operator.mul, '/': operator.truediv, '%': operator.mod} #supported operations
    while True:
      try:
        st = []
        for idx in Input.split():
          if idx in ops:
            y,x = st.pop(),st.pop()
            z = ops[idx](x,y)
          else:
            z = float(idx)
          st.append(z)
        assert len(st)<=1
        if len(st)==1: 
            print(int(st.pop()))
            break
        return st
            
            
      # for erronous input      
      except EOFError:  break
      except:  print('error')
```

```python
Input = input()
RPN(Input)
```
